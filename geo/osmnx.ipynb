# OSMnx

osmnx 是和 Open Street Map 搭配使用的，目前我主要使用它来获取行政单位的范围。
输入 OSM ID，返回 polygon

```python
import pandas as pd
import geopandas as gpd
import osmnx as ox
```

```python
places = pd.read_csv('../data/china_osmid.csv', sep='|', header=0)

places
```

获取范围时，OSM ID 前必须加上它们的类别：

- node (N)
- way (W)
- relation (R)

```python
osmids = ['R'+str(osmid) for osmid in places['osmid'].tolist()]

osmids
```

```python
places_df = ox.geocode_to_gdf(osmids, by_osmid=True)

places_df
```

展示获取的范围

```python
places_df.geometry[0]
```

```python
places_df.to_file('../data/places.geojson', driver='GeoJSON')
```
